```python
data1 = pd.read_excel('Keyword dummy data.xlsx')
data2 = pd.read_excel('Sponsored Products Advertised product report.xlsx')
data3 = pd.read_excel('keyword sheets.xlsx')
data4 = pd.read_csv('keywords_cleaned_patial.csv')
data1.info(), data2.info(), data3.info(), data4.info()
```

```python
set.intersection(set(data1.columns), set(data2.columns), set(data3.columns), set(data4.columns))
```

```python
common_features = set.intersection(set(data1.columns), set(data3.columns), set(data4.columns)) - set('Date')
```

```python
data = pd.concat([data1[common_features], data3[common_features], data4[common_features]]).drop_duplicates()
data.info()
```

```python
from numpy import inf

data.loc[data['Total Return on Advertising Spend (RoAS)'] == 0, 'Total Advertising Cost of Sales (ACoS)'] = None
data.loc[data['Total Return on Advertising Spend (RoAS)'] == None, 'Total Advertising Cost of Sales (ACoS)'] = None
data['Total Advertising Cost of Sales (ACoS)'] = 1 / data['Total Return on Advertising Spend (RoAS)']
data.loc[data['Total Advertising Cost of Sales (ACoS)'] == inf, 'Total Advertising Cost of Sales (ACoS)'] = None
data.info()
```

```python
data = data.drop(columns=['Total Advertising Cost of Sales (ACoS)'])
data
```

```python
num_features = ['Bid Value', 'Suggested lower bid range', 'Suggested upper bid range',
                'Impressions', 'Spend', 'Click-Thru Rate (CTR)', 
                'Total Return on Advertising Spend (RoAS)', 
                'Clicks', '14 Day Total Sales (₹)',
                '14 Day Conversion Rate', '14 Day Total Units (#)', '14 Day Total Orders (#)']

cat_features = ['Targeting', 'Match Type']

data['Bid Value'] = [float(str(val).split(' ')[-1]) for val in data['Bid Value']]
data['14 Day Total Sales (₹)'] = [float(str(val).replace(',', '').split(' ')[-1]) for val in data['14 Day Total Sales (₹)']]
data['Suggested lower bid range'] = [float(str(val).split(' ')[-1]) for val in data['Suggested lower bid range']]
data['Suggested upper bid range'] = [float(str(val).split(' ')[-1]) for val in data['Suggested upper bid range']]
data['Spend'] = [float(str(val).replace(',', '').split(' ')[-1]) for val in data['Spend']]
data = data[cat_features + num_features]
data.info()
```

```python
data = data[data['Click-Thru Rate (CTR)'].notna()]
data.head()
```

```python
data.to_csv('amazon_data_nan.csv', index=False)
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(drop='if_binary', sparse=False)
X_enc = enc.fit_transform(data[cat_features])
X_enc.shape
enc_columns = list(range(X_enc.shape[1]))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.neighbors import KNeighborsRegressor

targets = set(['Suggested lower bid range', 'Suggested upper bid range', '14 Day Conversion Rate', 'Total Return on Advertising Spend (RoAS)'])

X = data[set(num_features) - targets]
for c in enc_columns:
    X[str(c)] = X_enc[:, c]

for target in list(targets):
  print(target)
  y = data[target]

  y_train = y.dropna(inplace=False)

  not_null_idx = [i for i, val in enumerate(y.notnull()) if val]
  null_idx = [i for i, val in enumerate(y.notnull()) if not val]

  X_train = X.iloc[not_null_idx]
  print(X_train.shape)
  X_val = X_train[:500]
  y_val = y_train[:500]
  X_test = X.iloc[null_idx]
  y_test = y.iloc[null_idx]
  print(X_test.shape)
  knn = KNeighborsRegressor()
  knn.fit(X_train, y_train)
  preds = knn.predict(X_val)

  print('MAPE', mean_absolute_percentage_error(preds, y_val))
  print('R2', knn.score(X_val, y_val))

  preds = knn.predict(X_test)

  y_new = y
  y_new.iloc[null_idx] = preds
  print(sum(y_new.isna()))

  data_new = data[set(num_features) - targets]
  data_new[target] = y_new
  X[target] = y_new
```
